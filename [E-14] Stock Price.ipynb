# STEP0 : ARIMA모델에 대한 간략한 설명

`ARIMA모델`은 추세차분이 적용된 AR과 MA이 결합된 모델이다.
$$ \Delta^d Y_{t} = ARMA(p, q) $$

이때 I항을 제외한 `ARMA모델`은 `AR모델`과 `MA모델`의 합으로 나타난다.
- `AR모델` $ Y_{t} = \phi_{1}Y_{t-1} + \phi_{2}Y_{t-2} + ... + \phi{p}Y_{t-p} + e_{t} $ 
  - Auto Regression을 가정. 따라서 $Y_{t}$는 **자기자신**의 선형조합으로 결정된다고 보는 모델.
- `MA모델` $ Y_{t} = \theta_{1}e_{t-1} + \theta_{2}e_{t-2} + ... + \theta_{q}e_{t-q} + e_{t} $
  - Moving Average를 가정. 따라서 $Y_{t}$는 **백색잡음**의 선형조합으로 결정된다고 보는 모델.
- `ARMA모델`$ Y_{t} = \phi_{1}Y_{t-1} + \phi_{2}Y_{t-2} + ... + \phi{p}Y_{t-p} + \theta_{1}e_{t-1} + \theta_{2}e_{t-2} + ... + \theta_{q}e_{t-q} + e_{t} $ 
  - AR과 MA를 동시에 가정. 따라서 $Y_{t}$는 자기자신의 선형조합이며 백색잡음의 선형조합으로 결정된다고 보는 모델.

또 $Y_{t}$의 추세차분을 적용한 확률변수 $\Delta^d Y_{t}$의 `ARMA모델`은 $Y_{t}$의 `ARIMA모델`이 된다.
  - d = 1의 경우는 다음과 같이 쓸 수 있다: $Y_{t} - Y_{t-1} = \phi_{1}Y_{t-1} + ... + \phi_{p}Y_{t-p} + \theta_{1}e_{t-1} +  ... + \theta_{q}e_{t-q} + e_{t}$


또한 `ARMA모델`은 모두 $Y_{t}$의 Stationary를 가정하는데, 이는 차분, 로그변환 등으로 추세와 계절성을 제거한 상태를 의미한다. `ARIMA모델`에서는 $Y_{t}$가 추세차분을 통해 Stationary를 가질 경우 모델 자체에서 추세차분을 해준다.

> `ARIMA모델`은 `AR모델`과 `MA모델`의 결합이다. 또한 $Y_{t}$에 대한 추세차분을 내부적으로 적용해준다. 따라서 `ARIMA모델`에서 결정해주어야 할 모수는 p, q, d로 각각 AR, MA의 차수와 차분의 깊이로 결정된다. 

※ 논의를 따라가다보면 위 모델들이 모두 $Y_{t}$의 추세(Stationary한 부분)만을 설명하고 있는 것을 알 수 있다. 우리가 궁극적으로 알고 싶은 것은 이 시계열에 특정한 사건이 일어나서 실제로 상승하는지 감소하는지 일 것이다. 그렇다면 Unstationary한 부분에 있는 잔차를 설명하는 모델을 만들어야하고, 이것은 `ARIMAX모델`, `SARIMAX모델` 등에서 외생변수 X를 도입하여 해결한다고 한다. 향후 이러한 모델에 대한 공부를 이어가보자.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose

from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

from itertools import product
```

# STEP 3 : 시계열 안정성 분석
- 정성적 그래프 분석
- 정량적 Augmented Dicky-Fuller Test
- 시계열 분해(Time Series Decomposition)
- Residual 안정성 확인

```python
def plot_rolling_statistics(timeseries, window=12):
    
    rolmean = timeseries.rolling(window=window).mean()  # 이동평균 시계열
    rolstd = timeseries.rolling(window=window).std()    # 이동표준편차 시계열

     # 원본시계열, 이동평균, 이동표준편차를 plot으로 시각화해 본다.
    orig = plt.plot(timeseries, color='blue',label='Original')    
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import adfuller

def augmented_dickey_fuller_test(timeseries):
    # statsmodels 패키지에서 제공하는 adfuller 메소드를 호출합니다.
    dftest = adfuller(timeseries, autolag='AIC')  
    
    # adfuller 메소드가 리턴한 결과를 정리하여 출력합니다.
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
# 정성적 그래프 분석
plot_rolling_statistics(ts_log, window=12)
```

```python
#정량적 Augmented Dicky-Fuller Test
augmented_dickey_fuller_test(ts_log)
```

 - dickey fuller test의 귀무가설을 기각하지 못해 비정상성을 따르는 데이터라고 볼 수 있다. 따라서 시계열 분해를 시도한다.

```python
#시계열 분해 (Time Series Decomposition)
decomposition = seasonal_decompose(ts_log, model='multiplicative', period = 30) 

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
residual.dropna(inplace=True)
augmented_dickey_fuller_test(residual)
```

- 잔차는 dickey fuller test의 귀무가설을 기각하고 정상성을 만족한다. 따라서 잔차는 정상성을 따르는 백색잡음이다!

# STEP 4 : 학습, 테스트 데이터셋 생성

```python
train_data, test_data = ts_log[:int(len(ts_log)*0.9)], ts_log[int(len(ts_log)*0.9):]
plt.figure(figsize=(10,6))
plt.grid(True)
plt.plot(ts_log, c='r', label='training dataset')  # train_data를 적용하면 그래프가 끊어져 보이므로 자연스러운 연출을 위해 ts_log를 선택
plt.plot(test_data, c='b', label='test dataset')
plt.legend()
```

# STEP 5 : 적정 ARIMA 모수 찾기
- ACF, PACF 그려보기 -> p,q 구하기
- 차분 안정성 확인 -> d 구하기

```python
plot_acf(ts_log)   # ACF : Autocorrelation 그래프 그리기
plot_pacf(ts_log)  # PACF : Partial Autocorrelation 그래프 그리기
plt.show()
```

- PACF를 보고 AR 모형의 p를 결정한다 : 1이라고 볼 수 있음
- ACF를 보고 MA 모형의 q를 결정한다 : 명확히 결정할 수 없음. 그리드 서치를 이용해보자

```python
# 1차 차분 구하기
diff_1 = ts_log.diff(periods=1).iloc[1:]
diff_1.plot(title='Difference 1st')

augmented_dickey_fuller_test(diff_1)

# 혹시 필요한 경우 2차 차분 구하기
diff_2 = diff_1.diff(periods=1).iloc[1:]
diff_2.plot(title='Difference 2nd')

augmented_dickey_fuller_test(diff_2)
```

- `ARIMA모델`은 추세차분을 적용한 정상화 작업을 자동으로 진행해준다. 따라서 d만 결정해준채로 원 데이터인 ts_log를 입력으로 주면 ARMA모델을 뱉어준다. 이때, $d <= 2$를 이용한다.

# STEP 6 : ARIMA 모델 훈련과 테스트

- 여러 모형 중 Log Likelihood의 최대값을 갖는 파라미터를 찾아 기준으로 해보자.

```python
# ARIMA(p,d,q) 모델링
result = []
for p, d, q in product(range(1,2), range(1,3), range(1,10)):
    if (p == 0 & q == 0):
        continue
    model = sm.tsa.ARIMA(train_data, (p, d, q)).fit()
    try:
        result.append({"p": p, "d": d, "q": q, "LLF": model.llf, "AIC": model.aic, "BIC": model.bic})
    except:
        pass

#모형 최적모수 선택
result = pd.DataFrame(result)
display(result)


opt_ar = result.iloc[np.argmax(result['LLF']), 0]
opt_i = result.iloc[np.argmax(result['LLF']), 1]
opt_ma = result.iloc[np.argmax(result['LLF']), 2]
```

```python
plt.subplot(121)
result['LLF'][result['d'] == 1].plot()
plt.subplot(122)
result['LLF'][result['d'] == 2].plot()
```

- likelihood값을 보면 
  - d는 1인 경우가 전체적으로 크다.
  - q는 차수의 증가에 따라 큰 차이가 없음. 따라서 처음 증가하는 부분인 4 정도를 기준으로 한다.

```python
# ARMA 모델링
fitted_m = sm.tsa.ARIMA(train_data, (1,1, 1)).fit(disp=-1)
display(fitted_m.summary())
```

```python
# ARMA 모델링
fitted_m = sm.tsa.ARIMA(train_data, (1,1, 4)).fit(disp=-1)
display(fitted_m.summary())
```

- (1, 1, 4) 모델에서 p value 를 보면 AR의 첫째항과 MA의 첫째항이 유의미함을 알 수 있다.

```python
fitted_m.plot_predict()
```

```python
# Forecast : 결과가 fc에 담깁니다. 
fc, se, conf = fitted_m.forecast(len(test_data), alpha=0.05)  # 95% conf

# Make as pandas series
fc_series = pd.Series(fc, index=test_data.index)   # 예측결과
lower_series = pd.Series(conf[:, 0], index=test_data.index)  # 예측결과의 하한 바운드
upper_series = pd.Series(conf[:, 1], index=test_data.index)  # 예측결과의 상한 바운드

# Plot
plt.figure(figsize=(10,5), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, c='b', label='actual price')
plt.plot(fc_series, c='r',label='predicted price')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.10)
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(np.exp(test_data), np.exp(fc))
print('MSE: ', mse)

mae = mean_absolute_error(np.exp(test_data), np.exp(fc))
print('MAE: ', mae)

rmse = math.sqrt(mean_squared_error(np.exp(test_data), np.exp(fc)))
print('RMSE: ', rmse)

mape = np.mean(np.abs(np.exp(fc) - np.exp(test_data))/np.abs(np.exp(test_data)))
print('MAPE: {:.2f}%'.format(mape*100))
```

# 📕 회고
- 이번 노드는 주식 데이터를 시계열 분석을 통해 예측하는 내용이었다.
- 시계열 분석은 통계적 기반 탄탄한 분석방법으로 정확한 이해를 위해서는 많은 시간을 요구한다.
- 짧은 시간동안 시계열에 대한 이해를 향상시키기 위해 일반선형확률과정부터 적분선형확률과정까지 공부를 하였으나 노드에 추가적인 내용을 적용하기에는 시간이 부족했다.
- 그러나 p, d, q와 AR, I, MA의 관계를 어렴풋이 이해할 수 있었고, 이를 통해 이해를 기반으로 파라미터의 변주를 시도할 수 있었다.
- 비록 시계열분석은 RNN보다 성능은 떨어지지만 설명이 가능한 모델이라는 측면에서 강력함을 보인다. 또한 학습한 기간이 짧아 명확한 비전은 제시할 수 없지만, 설명 가능한 모델을 이해하기 위해서는 성능이 떨어지더라도 설명이 가능한 모델을 쌓아 올리는 방식의 학습방법이 필수적이라고 생각한다.
