# 1. ÏñºÍµ¥Ïù∏Ïãù Î∞è ÏûêÎ•¥Í∏∞ : MODULE face_recognition
- def face_recognition.load_image_file( ... )
- def face_recognition.face_locations( ... )
  - face_locations: Ïö∞ÏÉÅÎã® (a, b), Ï¢åÌïòÎã®(c, d) ÏñºÍµ¥Ïùò Ï¢åÌëúÎ•º Î¶¨Ïä§Ìä∏Î°ú Îã¥Í≥† ÏûàÎã§
  
- [ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ìï®Ïàò] def get_cropped_face(image_file) return cropped_face

## Ïù¥ÎØ∏ÏßÄÎ•º Í∞ÄÏ†∏ÏôÄÏÑú ÏûêÎ•¥Îäî ÌîÑÎ°úÏÑ∏Ïä§Î•º Ìï®ÏàòÎ°ú ÎßåÎì§Ïñ¥Î≥¥Ïûê

```python
def get_cropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
#     print(face_locations)  # Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÏñºÍµ¥ ÏòÅÏó≠Ïùò Ï¢åÌëúÎ•º Ï∂úÎ†•Ìï©ÎãàÎã§.
    a, b, c, d = face_locations[0] # Ï¢åÏÉÅÎã® Ï¢åÌëú, Ïö∞ÌïòÎã® Ï¢åÌëúe
    cropped_face = image[a:c,d:b,:] #(a, d) ~ (c, b)
    
    return cropped_face
```

## Ìï®Ïàòget_cropped_face ÌÖåÏä§Ìä∏
- [Trouble] face_recognition.face_locationsÏù¥ ÎêòÏßÄ ÏïäÎäî Î¨∏Ï†ú Î∞úÏÉù
  - ÏÇ∞Îì§.jpg : ÏñºÍµ¥ Í∏∞Ïö∏Ïñ¥Ïßê Î¨∏Ï†úÎ°ú ÏïàÎêòÎäî Í≤ÉÏúºÎ°ú ÏòàÏÉÅ
  - ‚Üí ÏÇ≠Ï†ú

# 2. FaceNet ÏñºÍµ¥ ÏûÑÎ≤†Îî© Î™®Îç∏ ÌôúÏö©ÌïòÍ∏∞
- def face_recognition.face_encodings( ... ) : RETURN list of 1Ï∞®Ïõê array, 128Ï∞®Ïõê Î≤°ÌÑ∞

- [ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ìï®Ïàò] def get_face_embedding(face) return face_recognition.face_encodings(face)

```python
# ÏñºÍµ¥ ÏòÅÏó≠ÏùÑ Í∞ÄÏßÄÍ≥† ÏñºÍµ¥ ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Î•º Íµ¨ÌïòÎäî Ìï®Ïàò
def get_face_embedding(cropped_face):
    return face_recognition.face_encodings(cropped_face)
```

## Ìï®Ïàòget_face_embedding ÌÖåÏä§Ìä∏

- [Trouble] face_recognition.face_encodingsÏúºÎ°ú ÏûÑÎ≤†Îî© ÎêòÏßÄ ÏïäÎäî Î¨∏Ï†ú Î∞úÏÉù

```python
error = []
for file in file_list:
    image_path = os.path.join(dir_path, file)
    face = get_cropped_face(image_path)   # ÏñºÍµ¥ ÏòÅÏó≠ÏùÑ Íµ¨ÌïòÎäî Ìï®Ïàò(Ïù¥Ï†Ñ Ïä§ÌÖùÏóêÏÑú Íµ¨ÌòÑ)
    try :
        embedding = get_face_embedding(face)
    except :
        error.append(file)
    else :
        if not embedding :
            error.append(file)
        else :
            pass
#             print(file + 'ÏÑ±Í≥µ')
if not error:
    print('Process succeeded')
```

## Ìé∏ÏùòÌï®Ïàòget_face_embedding_dict ÏûëÏÑ±
- def get_face_embedding_dict(dir_path) return embedding_dict
  - input: dir_path
  - output: Î™®Îì† Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùò ÏûÑÎ≤†Îî© dict {ÏÇ¨ÎûåÏù¥Î¶Ñ: ÏûÑÎ≤†Îî©Î≤°ÌÑ∞}

- [Trouble] jupyter dead kernel Î¨∏Ï†ú
  1. Î©îÎ™®Î¶¨Î•º Í≥ºÎèÑÌïòÍ≤å ÏÇ¨Ïö©ÌïòÎäî Î¨∏Ï†úÎ°ú, ÏÉùÏÑ±Îêú ÎîïÏÖîÎÑàÎ¶¨Î•º forÎ¨∏ÏúºÎ°ú Ï†ëÍ∑º Ïãú Ïª§ÎÑêÏù¥ Ï£ΩÎäî Î¨∏Ï†ú Î∞úÏÉù
     - generator comprehensionÏùÑ ÏÇ¨Ïö©Ìï¥ Î©îÎ™®Î¶¨ Î∂ÄÌïòÎ•º Ï§ÑÏó¨Î≥¥Í∏∞Î°ú Ìï®
     - biden.jpgÎ•º ÏÇ≠Ï†úÌïòÍ≥† Ïª§ÎÑêÏù¥ Ï£ΩÏßÄ ÏïäÏúºÎØÄÎ°ú generatorÎ•º ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏïÑÎèÑ ÏâΩÍ≤å Ï†ëÍ∑º Í∞ÄÎä•

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    
    embedding_dict = {}
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_cropped_face(img_path)
        embedding = get_face_embedding(face)
        if embedding:
#             print(f'{os.path.splitext(file)[0]}Ïùò ÏûÑÎ≤†Îî© ÏûÖÎãàÎã§.')
#             yield os.path.splitext(file)[0], embedding[0]
            #embeddingÏùÄ list of 1Ï∞®Ïõê arrayÏù¥ÎØÄÎ°ú Í∑∏ ÏõêÏÜåÎßå Í∫ºÎÇ¥Ïò¥
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
    return embedding_dict
```

```python
embedding_dict = get_face_embedding_dict(dir_path)
```

# 3. ÌöçÎìùÌïú ÏñºÍµ¥ Í∞Ñ Í±∞Î¶¨ Ï∏°Ï†ï : L2 distance
- L2 distance: Î≤°ÌÑ∞ a, bÏùò Ï∞®Ïóê ÎåÄÌïú p = 2 Ïù∏ norm
- [ÏÇ¨Ïö©Ïûê ÏÑ§Ï†ï Ìï®Ïàò] def get_distance(name1, name2) return Îëê Î≤°ÌÑ∞ Ï∞®Ïùò norm

```python
def get_L2_distance(name1, name2):
    import numpy as np
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)
```

# 4. ÎÇòÏôÄ Í∞ÄÏû• ÎãÆÏùÄ(= Í±∞Î¶¨Í∞Ä ÏßßÏùÄ) ÏñºÍµ¥ Ï∞æÏïÑÎ≥¥Í∏∞

>1. def get_cropped_face(image_file) :
	 Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏóêÏÑú ÏñºÍµ¥ ÏòÅÏó≠ÏùÑ Í∞ÄÏ†∏Ïò§Îäî Ìï®Ïàò

     ÏñºÍµ¥ÏòÅÏó≠ÏúºÎ°úÎ∂ÄÌÑ∞ ÏñºÍµ¥ ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Î•º Íµ¨ÌïòÎäî Ìï®Ïàò
>3. def get_face_embedding_dict(dir_path) : 
     ÎîîÎ†âÌÜ†Î¶¨ ÏïàÏóê ÏûàÎäî Î™®Îì† Ïù¥ÎØ∏ÏßÄÏùò ÏûÑÎ≤†Îî© ÎîïÏÖîÎÑàÎ¶¨Î•º Íµ¨ÌïòÎäî Ìï®Ïàò
>4. def get_distance(name1, name2) : 
     Îëê Ïù¥ÎØ∏ÏßÄ(ÏÇ¨Îûå Ïù¥Î¶Ñ) ÏÇ¨Ïù¥Ïùò ÏûÑÎ≤†Îî© Î≤°ÌÑ∞ Í±∞Î¶¨Î•º Íµ¨ÌïòÎäî Ìï®Ïàò
- Íµ¨ÌïòÎ†§Îäî Ìï®Ïàò
  - input: name / output: [Í±∞Î¶¨Ï†ïÎ≥¥, ... ]


- [ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ìï®Ïàò] def get_sort_key_func(name1): return get_L2_distance_from_name1
  - name1Í≥º name2Ïùò Í±∞Î¶¨Î•º ÎπÑÍµêÌïòÎäî ÌÅ¥Î°úÏ†ÄÎ•º ÏÉùÏÑ±

```python
def get_sort_key_func(name1):
    def get_L2_distance_from_name1(name2):
        return get_L2_distance(name1, name2)
    return get_L2_distance_from_name1
```

- ÏÇ¨Ïö©Ïûê Ìé∏Ïùò(Ïù¥Î¶ÑÎßå Ï†ÑÎã¨)ÏùÑ Ïú†ÏßÄÌïòÎ©¥ÏÑú Í∑∏Î¶ºÏùÑ Í∑∏Î¶¨Í∏∞ ÏúÑÌï¥ ÌôïÏû•ÏûêÎ•º Í∞ÄÏ†∏Ïò§Îäî Ïú†Ìã∏Î¶¨Ìã∞ Ìï®ÏàòÎ•º ÏûëÏÑ±ÌïúÎã§

```python
def get_extension(name):
    file_list = os.listdir(dir_path)
    for src_name, ext in list(map(lambda x : x.split('.'), file_list)):
        if name == src_name:
            return '.' + ext
    else :
        print("Ï°¥Ïû¨ÌïòÏßÄ ÏïäÎäî ÌååÏùºÏûÖÎãàÎã§!")
```

```python
def get_nearest_face(name, top=5):
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))

    #ÎãÆÏùÄ ÏÇ¨ÎûåÏùò ÏñºÍµ¥ÏùÑ ÌëúÌòÑÌïòÍ∏∞ ÏúÑÌï¥ ÏûëÏÑ±Ìïú Î∂ÄÎ∂ÑÏûÑ
    plt.rcParams["figure.figsize"] = (12,8)
    fig, axes = plt.subplots(nrows = top//3 + 1, ncols = 3 )
    axes = axes.flatten()
    image = img.imread(os.path.join(dir_path, name + get_extension(name) ))
    axes[0].imshow(image)

    for i in range(top+1):
        if i == 0 :   # Ï≤´Î≤àÏß∏Î°ú ÎÇòÏò§Îäî Ïù¥Î¶ÑÏùÄ ÏûêÍ∏∞ ÏûêÏã†Ïùº Í≤ÉÏù¥ÎØÄÎ°ú Ï†úÏô∏Ìï©ÏãúÎã§. 
            continue
        if sorted_faces[i]:
            compname = sorted_faces[i][0]
            distance = sort_key_func(sorted_faces[i][0])
            print(f'ÏàúÏúÑ {i} : Ïù¥Î¶Ñ({compname}), Í±∞Î¶¨({distance})')
            
            image = img.imread(os.path.join(dir_path, compname + get_extension(compname) ))
            axes[i].imshow(image)
```
